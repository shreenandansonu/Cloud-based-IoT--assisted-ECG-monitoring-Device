```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your CSV file into a pandas DataFrame
# Replace 'your_data.csv' with the actual path to your CSV file
df = pd.read_csv('shree1000.csv')

# Choose the column you want to apply the moving average filter to
column_to_filter = 'volt'  # Replace 'ColumnName' with your column name

# Set the window size for the moving average
window_size = int(input(print("Window Size: ")))  # Adjust the window size as needed

# Apply the moving average filter using the rolling() function
filtered_data = df[column_to_filter].rolling(window=window_size).mean()

# Create a plot to visualize the original and filtered data
plt.figure(figsize=(20,5))

# Plot the original data
plt.plot(df['time'], df[column_to_filter], label='Original Data', color='b')

# Plot the filtered data
plt.plot(df['time'], filtered_data, label='Moving Average', color='r')

# Customize the plot
plt.title('Moving Average Filter')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
name=(input(print("Save image as: ")))
plt.savefig(''+name+'.png')
# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your ECG data
file_name = 'shree1000.csv'
df = pd.read_csv(file_name)

# Define a threshold value (adjust as needed)
threshold = 2.8  # Set your desired threshold value

# Initialize variables to store peak indices and values
peak_indices = []
peak_values = []

# Iterate through the data points to detect peaks
for i in range(1, len(df['volt']) - 1):
    if df['volt'][i] > threshold and df['volt'][i] > df['volt'][i - 1] and df['volt'][i] > df['volt'][i + 1]:
        peak_indices.append(i)
        peak_values.append(df['volt'][i])

# Plot the ECG data
plt.figure(figsize=(20, 5))
plt.plot(df['time'], df['volt'], color='red', label='ECG Data')

# Plot the detected peaks
plt.plot(df['time'][peak_indices], peak_values, 'o', markersize=5, color='blue', label='Peaks')

plt.title('ECG Data with Peaks Detected')
plt.xlabel('Time in seconds')
plt.ylabel('Voltage amplified in 0-5 V')
plt.grid(True)
plt.legend()
name=(input(print("Save image as: ")))
plt.savefig(''+name+'.png')
plt.show()
# Display the coordinates of the detected peaks
peak_coordinates = [(df['time'][idx], value) for idx, value in zip(peak_indices, peak_values)]
print("Detected Peak Coordinates:")
for coordinate in peak_coordinates:
    print(coordinate)

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your ECG data
file_name = 'shree1000.csv'
df = pd.read_csv(file_name)

# Define a threshold value (adjust as needed)
threshold = 2.8  # Set your desired threshold value

# Initialize variables to store peak indices and values
peak_indices = []
peak_values = []

# Iterate through the data points to detect peaks
for i in range(1, len(df['volt']) - 1):
    if df['volt'][i] > threshold and df['volt'][i] > df['volt'][i - 1] and df['volt'][i] > df['volt'][i + 1]:
        peak_indices.append(i)
        peak_values.append(df['volt'][i])

# Calculate the time interval between subsequent peaks
time_intervals = [df['time'][peak_indices[i+1]] - df['time'][peak_indices[i]] for i in range(len(peak_indices) - 1)]

# Calculate heart rate (beats per minute) for each peak
heart_rates = [60 / time_interval for time_interval in time_intervals]

# Plot the heart rate
plt.figure(figsize=(20, 5))
plt.plot(df['time'][peak_indices[:-1]], heart_rates, color='green', marker='o', linestyle='-', markersize=5, label='Heart Rate (BPM)')
plt.title('Heart Rate Plot')
plt.xlabel('Time in seconds')
plt.ylabel('Heart Rate (BPM)')
plt.grid(True)

# Calculate the mean heart rate
mean_heart_rate = sum(heart_rates) / len(heart_rates)
print("Mean Heart Rate:", mean_heart_rate, "BPM")
name=(input(print("Save image as: ")))
plt.savefig(''+name+'.png')
# Add a horizontal line for the mean heart rate
plt.axhline(mean_heart_rate, color='blue', linestyle='--', label=f'Mean Heart Rate: {mean_heart_rate:.2f} BPM')
plt.legend()
plt.show()
```
